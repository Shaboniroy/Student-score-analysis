# **DROP** **UNNAMED** **COLUMN**

```python
dataframe = dataframe.drop("Unnamed: 0", axis = 1)
print(dataframe.head())
```

# **1)Gender Distribution**

```python
plt.figure(figsize = (5,5))
ax = sns.countplot(data = dataframe, x = "Gender")
plt.title("Gender Distribution")
ax.bar_label(ax.containers[0])
plt.show()
```

**1) From the above chart we have analyzed that the number of females is more as compared to the number of males.**

# **2)Relationship between Parent's education and Student's score**

```python
gb = dataframe.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
gb
```

```python
plt.figure(figsize= (4,5))
sns.heatmap(gb, annot = True)
plt.title("Relationship between Parent's education and Student's score")
plt.show()
```

**2) From the above chart we have analyzed that the parent's education have a good impact on their children scores**

# **3) Relationship between Parent's martial status and Student's score**

```python
gb1 = dataframe.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
gb1
```

```python
plt.figure(figsize=(4,5))
sns.heatmap(gb1,annot = True)
plt.title("Relationship between Parent's Martial Status and Student's Score")
plt.show()
```

**3) From the above chart we concluded that parent's martial status has negligible effect on the student's score.**

# **5) Distribution of Ethnic Groups**

```python
GroupA = dataframe.loc[(dataframe['EthnicGroup'] == "group A")].count()
GroupB = dataframe.loc[(dataframe['EthnicGroup'] == "group B")].count()
GroupC = dataframe.loc[(dataframe['EthnicGroup'] == "group C")].count()
GroupD = dataframe.loc[(dataframe['EthnicGroup'] == "group D")].count()
GroupE = dataframe.loc[(dataframe['EthnicGroup'] == "group E")].count()
```

```python
l = ["group A", "group B", "group C", "group D", "group E"]
mlist = [GroupA["EthnicGroup"], GroupB["EthnicGroup"], GroupC["EthnicGroup"], GroupD["EthnicGroup"], GroupE["EthnicGroup"]]
plt.pie(mlist, labels = l, autopct = "%1.2f%%")
plt.title("Distribution of Ethnic Groups")
plt.show()
```

**6) Group C is the highest with 31.98% among all other Ethnic Groups**

# **7)Relationship between Sports practice and Student's scores**
